```python
class GradientDescent(object):
    def __init__(self, alpha=0.01, number_of_dataset=10):
        self.alpha = alpha
        self.number_of_dataset = number_of_dataset
        
    def fit(self, X_axis, Y_axis):
        self.weight = np.zeros(2)
        self.errors = []
        
        for i in range(self.number_of_dataset):
            errors = 0
            for j in range(X_axis.shape[0]):
                self.weight[1:] += self.alpha*X_axis[j]*(Y_axis[j] - self.weight[0] - self.weight[1]*X_axis[j])
                self.weight[0] += self.alpha*(Y_axis[j] - self.weight[0] - self.weight[1]*X_axis[j])
                errors += 0.5*(Y_axis[j] - self.weight[0] - self.weight[1]*X_axis[j])**2
            self.errors.append(errors)
        return self
    
    def prediction(self, X_axis):
        return self.weight[0] + self.weight[1]*X_axis
    
    def square(self,X_axis, Y_axis):
        return 1-((self.prediction(X_axis) - Y_axis)**2).sum()/((Y_axis - np.mean(Y_axis))**2).sum()
        
```

```python
plt.scatter(X_train, Y_train, color = 'blue')
plt.plot(X_train, GradientDescent(alpha=0.01, number_of_dataset=100).fit(X_train,Y_train).prediction(X_train), color ="yellow")
plt.title("Temperature VS Humidity (GradientDescent)")
plt.xlabel("Temperature")
plt.ylabel("Humidity")
plt.show()
```

# Question 8

Please cluster the following points, (18, 10), (21, 11), (22, 22), (24, 15), (26, 12), (26, 13), (27, 14), (30, 33), (31, 39), (35, 37), (39, 44), (40, 27), (41, 29), (42, 20), (44, 28), (46, 21), (47, 30), (48, 31), (49, 23), (54, 24) use the numbers as pairs of x and y values which represent their locations. All distances are measured with Euclidean distance.

A) Use the k-means algorithm, with (24, 15), (30, 33), and (54, 24) as the initial cluster centers, process the dataset until convergence, choose a method of your choice to evaluate the cluster result quality.

```python
def euclidean_distance(point, center):
    return np.sqrt(np.sum((point - center)**2))

def label_cluster(distance, point, center):
    minimum = min(distance)
    return [minimum, point, center[minimum]]

def new_center(cluster_label, center):
    return np.array(cluster_label + center)/2

def create_center():
    center = []
    center.append([24, 15])
    center.append([30, 33])
    center.append([54, 24])
    return np.array(center)
```

```python
def k_means(point, center, iteration = 100):
    label = []
    cluster = []
    
    for i in range(0, iteration):
        for j in range(0, len(point)):
            distance = {}
            for l in range(0, len(center)):
                distance[l] = euclidean_distance(point[j], center[l])
            label = label_cluster(distance, point[j], center)
            center[label[0]] = new_center(label[1], center[label[0]])
            if i == (iteration - 1):
                cluster.append(label)
    return [cluster, center]

def print_data(data_points):
    print("k-Means Clustering Algorithm:")
    for intergrate in data_points[0]:
        print("Point:", intergrate[1])
        print("Cluster:", intergrate[0],"\n")
    print("Center Point:", data_points[1],"\n")
```

```python
def get_data():
    data = []
    data.append([18,10])
    data.append([21,11])
    data.append([22,22])
    data.append([24,15])
    data.append([26,12])
    data.append([26,13])
    data.append([27,14])
    data.append([30,33])
    data.append([31,39])
    data.append([35,37])
    data.append([39,44])
    data.append([40,27])
    data.append([41,29])
    data.append([42,20])
    data.append([44,28])
    data.append([46,21])
    data.append([47,30])
    data.append([48,31])
    data.append([49,23])
    data.append([54,24])
    data = np.asarray(data)
    print(data)
    center = create_center()
    [cluster, new_center] = k_means(data, center, iteration = 100)
    print_data([cluster, new_center])
    
get_data()
```

```python
def Single_Complete_Average_Link(point, center, linkage_type, iteration = 100):
    label = []
    cluster = []
    
    if(linkage_type == "Single" or linkage_type == "single"):
        for i in range(0, iteration):
            for j in range(0, len(point)):
                distance = {}
                for l in range(0, len(center)):
                    distance[l] = euclidean_distance(point[j], center[l])
                label = label_cluster_single(distance, point[j], center)
                center[label[0]] = new_center(label[1], center[label[0]])
                if i == (iteration - 1):
                    cluster.append(label)
        return [cluster, center]
    
    elif(linkage_type == "Average" or linkage_type == "average"):
        for i in range(0, iteration):
            for j in range(0, len(point)):
                distance = {}
                for l in range(0, len(center)):
                    distance[l] = euclidean_distance(point[j], center[l])
                label = label_cluster(distance, point[j], center)
                center[label[0]] = new_center(label[1], center[label[0]])
                if i == (iteration - 1):
                    cluster.append(label)
        return [cluster, center]
    
    elif(linkage_type == "Complete" or linkage_type == "complete"):
        for i in range(0, iteration):
            for j in range(0, len(point)):
                distance = {}
                for l in range(0, len(center)):
                    distance[l] = euclidean_distance(point[j], center[l])
                label = label_cluster_complete(distance, point[j], center)
                center[label[0]] = new_center(label[1], center[label[0]])
                if i == (iteration - 1):
                    cluster.append(label)
        return [cluster, center]
    
    elif(linkage_type == "Centroid" or linkage_type == "centroid"):
        for i in range(0, iteration):
            for j in range(0, len(point)):
                distance = {}
                for l in range(0, len(center)):
                    distance[l] = euclidean_distance(point[j], center[l])
                label = label_cluster(distance, point[j], center)
                center[label[0]] = new_center(label[1], center[label[0]])
                if i == (iteration - 1):
                    cluster.append(label)
        return [cluster, center]
```
